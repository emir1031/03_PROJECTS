# Why are RNNS good for NLP?<a id="1"> </a><br><br><span style="color:BlueViolet">Neden RNN modeller NLP için iyidir?</span>

RNNs effectively have an internal **memory** that allows the previous inputs to affect the subsequent predictions. It's much easier to predict the next word in a sentence with more accuracy, **if you know what the previous words were**.
<br><br>
<span style="color:BlueViolet">RNN'ler, önceki girdilerin sonraki tahminleri etkilemesine izin veren etkin bir **dahili belleğe** sahiptir. **Önceki kelimelerin ne olduğunu biliyorsanız**, bir cümledeki bir sonraki kelimeyi daha doğru bir şekilde tahmin etmek çok daha kolaydır.</span>

# Which activation function is used in RNN models and Why?<a id="2"> </a><br><br><span style="color:BlueViolet">RNN modellerde Hangi aktivasyon fonksiyonu kullanılır ve Neden?</span>

When vectors pass through a neural network, they undergo a number of changes as a result of various mathematical operations. As a result of these processes, some values become very large and some values become insignificant. To prevent this, TanH activation is used. Thus, all vector values remain between -1 and +1.

The TanH activation function produces a value between -1 and +1 as an output, regardless of the input. In other words, it makes the weighting of our vectors proportionally. It does not give much weight to one vector and less to the other. Maintains weightings of tokens/ words/ inputs.
<br><br>
<span style="color:BlueViolet">Vektörler bir sinir ağından geçtiğinde, çeşitli matematik işlemler sonucunda bir takım değişikliklere maruz kalırlar. Bu işlemler sonucunda bazı değerlerin çok büyük, bazı değerlerin ise önemsiz hale gelmesi söz konusudur. Bunu engellemek için tanh aktivasyonu kullanılır. Böylece tüm vektör degerleri -1 ile +1 degerleri arasında kalır.</span>

<span style="color:BlueViolet">TanH aktivasyon fonksiyonu, girdi ne olursa olsun çıktı olarak -1 ve +1 arasında değer üretir. Yani vektörlerimizin ağırlıklandırılmasını oransal olarak yapar. Bir vektöre fazla ve diğerine daha az bir ağırlık vermez. Tokenların/ keimelerin/ girdilerin ağırlıklandırmalarını korur.</span>

# <a id="3"> </a>Why do we use LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Units)<br><br><span style="color:BlueViolet">LSTM (Uzun Kısa Süreli Bellek) ve GRU (Kapılı Tekrarlayan Birimler) kullanmamızın nedeni nedir?</span>

**Long Short Term Memory Network** is an advanced RNN, a sequential network, that allows information to **persist**. It is capable of **handling the vanishing gradient problem** faced by RNN. A recurrent neural network is also known as RNN is used for **persistent memory**.

- In RNNs, when a sentence was given as input, the beginning of the sentence (ie document) was forgotten as one moved towards the last word/token.
- **LSTM and GRU** models were developed to overcome this problem.
- The workflow of GRU is same as RNN but the difference is in the operations inside the GRU unit. There are **differences in its architecture**.
- The difference between LSTM and normal RNN; LSTM has **cell state**. It is a layer that holds cell state information.
- For more detailed information, I recommend a [Medium article that may be useful to you.](https://medium.com/analytics-vidhya/rnn-vs-gru-vs-lstm-863b0b7b1573)

<span style="color:BlueViolet">**Uzun Kısa Süreli Bellek Ağı (LSTM)**, bilginin **kalıcı** olmasını sağlayan sıralı bir ağ olan gelişmiş bir RNN'dir. RNN'nin karşılaştığı **kaybolan/ vanishing gradyan problemini** çözebilir. **Kalıcı bellek** için kullanılan RNN olarak da bilinen tekrarlayan bir sinir ağı.</span>

- <span style="color:BlueViolet">RNN'lerde bir cümle girdi olarak verildiğinde, son kelime/token'a doğru gelindikçe, cümlenin (yani document'in) başı unutuluyordu.</span>
- <span style="color:BlueViolet">Bu problemin giderilmesi için **LSTM ve GRU** modelleri geliştirildi.</span>
- <span style="color:BlueViolet">GRU'nun iş akışı, RNN ile aynıdır ancak fark, GRU birimi içindeki işlemlerdedir. **Mimarisini içerisinde farklılıklar** vardır.</span>
- <span style="color:BlueViolet">LSTM ile normal RNN'in farkı; LSTM'in **cell state**'e sahip olmasıdır. Hücre durum bilgisini tutan bir katmandır.</span>
- <span style="color:BlueViolet">Daha detaylı bilgi için sizlere faydalı olabilecek bir</span> [Medium makalesi öneriyorum.](https://medium.com/analytics-vidhya/rnn-vs-gru-vs-lstm-863b0b7b1573)

# The Architecture of LSTM<a id="4"> </a><br><br><span style="color:BlueViolet">LSTM Mimarisi</span>

- **Forget Gate :<a id="4.1"> </a>** This gate determines whether the information is stored or not. The sigmoid function is used for this operation. The closer the value is to zero, the more likely it is to be forgotten, and the closer to one, the more likely it will be remembered. It determines the patterns that the cell state should keep in its memory for a long time or that should be removed.
- <span style="color:BlueViolet">**Forget Gate :** Bu kapı bilgilerin saklanıp saklanılmayacağını belirler. Bu işlem için sigmoid fonksiyonu kullanılır. Değer sıfıra ne kadar yakınsa unutulma olasılığı o kadar yüksek ve bire ne kadar yakınsa akılda kalma olasılığı o kadar yüksek olacaktır. Cell state'in uzun sürede hafızasında durması gereken ya da çıkarılması gereken kalıpları belirler.</span>
- **Input Gate :<a id="4.2"> </a>** In the first stage, the data coming to the hidden state. Then the data is first subjected to the Sigmoid and then the TanH process. The sigmoid function actually decides whether the input should pass or not. Since the data passing through the sigmoid function will be between 0 and 1, the weights of these passing data pass through the TanH function and determine whether the data is added to the cell state.
- <span style="color:BlueViolet">**Input Gate :** İlk aşamada gizli durumda bulunan mevcut giriş (hidden state)'e gelen veri , önce Sigmoid ve ardından TanH işlemine tabi tutulur. Sigmoid fonksiyonu aslında girdinin geçip geçmeyeceğine karar verir. Sigmoid fonksiyonundan geçen veriler 0 ile 1 arasında olacağından, bu geçen verilerin ağırlıkları TanH fonksiyonundan geçer ve verinin hücre durumu (cell state)'na eklenip eklenmeyeceğini belirler. </span>
- **Cell State :<a id="4.3"> </a>** It is the state of the cell. It carries and preserves relevant information throughout the training period. In this way, previous information is not forgotten and the model can make more accurate predictions. In short, we can say that the cell state is the memory of the model/neural network.
- <span style="color:BlueViolet">**Cell State :** Hücrenin durumudur. Dizi boyunca ilgili bilgileri taşır ve korur. Bu sayede önceki bilgiler unutulmamakta ve model daha doğru tahminler yapabilmektedir. Kısaca cell state, modelin/ sinir ağının hafızasıdır da diyebiliriz.</span>
- **In short**;
- <span style="color:BlueViolet">**Kısacası**;</span>
    - **Hidden state =** Working memory capability that carries information from immediately previous events and overwrites at every step uncontrollably. Present at RNNs and LSTMs
    - <span style="color:BlueViolet">**Hidden state =** Hemen önceki olaylardan gelen bilgileri taşıyan ve her adımda kontrolsüz bir şekilde üzerine yazan çalışma belleği özelliği. RNN'lerde ve LSTM'lerde bulunur.</span>
    - **Cell state =** long term memory capability that stores and loads information of not necessarily immediately previous events. Present in LSTMs
    - <span style="color:BlueViolet">**Cell state =** hemen önceki olayların hemen olması gerekmeyen bilgileri depolayan ve yükleyen uzun süreli bellek yeteneği. LSTM'lerde mevcut.</span>

# What is GRU (Gated Recurrent Units)?<a id="6"> </a><br><br><span style="color:BlueViolet">GRU (Kapılı Tekrarlayan Birimler) Nedir?</span>



It is quite similar to LSTM. There are 2 gates. Update Gate and Reset Gate. In GRU, Hidden and Cell state are combined in a single layer.<br>
<span style="color:BlueViolet">LSTM ile oldukça benzerdir. 2 gate vardır. Update Gate ve Reset gate. GRU da Hidden ve Cell state tek bir katmanda birleştirilmişir.</span>

**Update Gate :<a id="6.1"> </a>** Update gate, it acts similar to an LSTM's forget and input gate. It decides what information to discard and what new information to add.<br>
<span style="color:BlueViolet">**Update Gate :**  Güncelleme kapısıdır, bir LSTM'nin forget ve input kapısına benzer şekilde hareket eder. Hangi bilgilerin atılacağına ve hangi yeni bilgilerin ekleneceğine karar verir.</span>

**Reset Gate :<a id="6.2"> </a>** It is the reset gate. It is another gate used to decide how much past knowledge will be forgotten. Like forget gate.<br>
<span style="color:BlueViolet">**Reset Gate :** Sıfırlama kapısıdır. Ne kadar geçmiş bilginin unutulacağına karar vermek için kullanılan başka bir kapıdır. Forget gate gibi yani.</span>

Transactions take longer in LSTM. GRU is faster in this regard.<br>
<span style="color:BlueViolet">LSTM de işlemler daha uzun sürmektedir. GRU bu konuda daha hızlıdır. </span>

Although GRU is faster than LSTM, both LSTM and GRU give very good results.<br>
<span style="color:BlueViolet">GRU, LSTM'den daha hızlı olmasına rağmen, hem LSTM hem de GRU çok iyi sonuçlar verir.</span>

The latest developed model is the GRU model. Which one to use is up to you.<br>
<span style="color:BlueViolet">En son geliştirilen model GRU modeldir. Hangisini kullanmak size kalmıştır. </span>
