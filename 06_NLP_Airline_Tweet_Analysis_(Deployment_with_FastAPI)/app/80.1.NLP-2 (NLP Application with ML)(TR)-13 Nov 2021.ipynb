## Cleaning Data

```python
# nltk.download('stopwords')
```

```python
# nltk.download('wordnet')
```

```python
stop_words = stopwords.words('english')
```

```python
def cleaning(data):
    
    #1. Tokenize
    text_tokens = word_tokenize(data.replace("'", "").lower())  # Yorumun olumsuz haliyle de olsa kalması için  ilave bir yöntem olarak kullanıldı.
    
    #2. Remove Puncs
    tokens_without_punc = [w for w in text_tokens if w.isalpha()]
    
    #3. Removing Stopwords
    tokens_without_sw = [t for t in tokens_without_punc if t not in stop_words]
    
    #4. lemma
    text_cleaned = [WordNetLemmatizer().lemmatize(t) for t in tokens_without_sw]
    
    #joining
    return " ".join(text_cleaned)
```

- iki tane def te aynı, aşağıdaki bir iinceki derste gösterilendir.
- yukarıdaki def fonksiyonunda ayraçları kaldırma işlemi gerçekleştirdik. upgrade ettik.
    - 1. ayraçlar kaldırıldı ve küçük harfe dönüştürüldü.
    - 2. noktalama işaretler ve numaralardan arındırılma işlemi yapıldı
    - 3. token without punc içindeki tüm tokenleri al ve içinden stop words olmayanlar ile yola devam et.
    - 4. stop word lerden arındırılmış tokenlerin lemma yani köklerine gittik.
    - 5. text imizi birleştirdik.

```python
" I don't like".replace("'", "")
```

```python
df2["text"] = df2["text"].apply(cleaning)
df2["text"].head()
```

## Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

- benim corpusum (datasetim) içerisindeki geçen bütün unique tokenlerimi sayan bir işlemdir.

```python
vectorizer = CountVectorizer()
X_train_count = vectorizer.fit_transform(X_train)
X_test_count = vectorizer.transform(X_test)
```

- **SAM**
- scale işleminde yaptığımız gibi, leakage önlemek için train --> fit_transform, test --> transform
- fit_transform dediğimiz için unique olan tüm kelimeleri sayıyor ve array e dönüştürüyor
- test te geçen bir kelime train de geçmiyorsa model o kelimeyi yok sayar, görmez.
- model ne kadar fazla kelime içeriyorsa o kadar bu handikapın önüne geçiyor.

- (X_train_count = vectorizer.fit_transform(X_train)) --> unique olan tüm kelimeleri sayıyor ve array e dönüştürüyor
- (X_test_count = vectorizer.transform(X_test)) --> X_test te geçen bir token, eğitildiği set yani x_train de geçmiyorsa onu almaz

## Model Comparisons - Vectorization

```python
from sklearn.metrics import plot_confusion_matrix,classification_report, f1_score, recall_score
```

```python
def eval(model, X_train, X_test):
    y_pred = model.predict(X_test)
    y_pred_train = model.predict(X_train)
    print("Test_Set")
    print(classification_report(y_test,y_pred))
    print("Train_Set")
    print(classification_report(y_train,y_pred_train))
    plot_confusion_matrix(model,X_test, y_test)
```

## TF-IDF

https://mdurmuss.github.io/tf-idf-nedir/

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tf_idf_vectorizer = TfidfVectorizer()
X_train_tf_idf = tf_idf_vectorizer.fit_transform(X_train)  # bütün unique token ve onların geçtiği satırları alıyor ve sayıyor
X_test_tf_idf = tf_idf_vectorizer.transform(X_test)
```

- TF = KELİMEM / TOPLAM KELİME SAYISI
- IDF = CORPUS TAKİ TOPLAM DOKÜMAN SAYISI / KELİMEMİN GEÇTİĞİ DOKÜMAN SAYISI

- tf_idf_vectorizer = TfidfVectorizer()
- X_train_tf_idf = tf_idf_vectorizer.fit_transform(X_train)
- X_test_tf_idf = tf_idf_vectorizer.transform(X_test)
     - tf_idf: fit deyince X_traindeki tüm unique tokenleri tespit ediyor ve bütün satırlarda (dokümanlarda) o tokenin geçme sıklığını da sayıyor.
     - count vectorizer da sadece unique tokenleri tespit ediyordu burda dokümanlardaki tekrar sayısı da işin içinde..
     - transform deyince bunları sayıyor ve fonksiyonu uygulayıp tf-idf katsayılarını buluyor.

## Cross Val

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score
custom_scorer = {'accuracy': make_scorer(accuracy_score),
                 'precision-neg': make_scorer(precision_score, average=None, labels=["negative"]),
                 'recall-neg': make_scorer(recall_score, average=None, labels = ["negative"]),
                 'f1-neg': make_scorer(f1_score, average=None, labels = ["negative"]),
                 
                 'precision-neutral': make_scorer(precision_score, average=None, labels=["neutral"]),
                 'recall-neutral': make_scorer(recall_score, average=None, labels = ["neutral"]),
                 'f1-neutral': make_scorer(f1_score, average=None, labels = ["neutral"]),
                 
                 'precision-pos': make_scorer(precision_score, average=None, labels=["positive"]),
                 'recall-pos': make_scorer(recall_score, average=None, labels = ["positive"]),
                 'f1-pos': make_scorer(f1_score, average=None, labels = ["positive"]),
                 }

for i, j  in custom_scorer.items():
    model = LogisticRegression(C=0.4, max_iter=1000)
    scores = cross_val_score(model, X_train_tf_idf, y_train, cv = 10, scoring = j, n_jobs = -1).mean()
    print(f" {i:20} score : {scores}\n")
```
